```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/95onu/Downloads/asset-v1_Kodlasam+MS01+2023_06+type@asset+block@nobel.csv')
```

```python
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
df['age'] = df['year'].dt.year - df['birth_date'].dt.year
```

```python
categories = df['category'].unique()
```

```python
df['year'] = df['year'].dt.year
```

```python
sns.set(style="whitegrid")
```

```python
for category in categories:
    category_data = df[df['category'] == category]
    
    g = sns.FacetGrid(category_data, col="sex", hue="sex", height=6, aspect=1.5, palette="Set1")
    g.map(sns.scatterplot, "year", "age", alpha=.7)
    g.add_legend()
    g.set_axis_labels("Yıl", "Yaş")
    g.set_titles(col_template="{col_name}")
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(f"{category} Nobel Ödülü İçin Yıllar İçinde Yaş ve Cinsiyet Dağılımı", size=16)
    plt.show()
```

# 7. soru: 1947-1991 yılları arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.(Her kategori için ayrı bir grafik olması istenmektedir)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1947_1991 = df[(df['year'] >= pd.Timestamp(1947, 1, 1)) & (df['year'] <= pd.Timestamp(1991, 12, 31))]
```

```python
category_counts = df_1947_1991['category'].value_counts()
```

```python
categories = df_1947_1991['category'].unique()
for category in categories:
    if category == 'Peace':
        country_counts = df_1947_1991[df_1947_1991['category'] == category]['birth_country'].value_counts()
    elif category == 'Literature':
        country_counts = df_1947_1991[df_1947_1991['category'] == category]['birth_country'].value_counts()
    else:
        country_counts = df_1947_1991[df_1947_1991['category'] == category]['organization_country'].value_counts()

    plt.figure(figsize=(12, 6))
    
    if len(country_counts) == 0:
        print(f"{category} kategorisi için veri bulunamadı.")
        continue

    cmap = "viridis"
    sns.barplot(x=country_counts.index, y=country_counts.values, palette=cmap)
    
    plt.xticks(rotation=90)
    plt.xlabel('Ülke')
    plt.ylabel('Ödül Sayısı')
    plt.title(f'{category} Kategorisi için 1947-1991 Yılları Arasında Nobel Ödülü Kazananların Ülkelere Göre Dağılımı')
    plt.show()
```

# 8. Soru:Kimya, Edebiyat, Barış, Fizik ve Tıp kategorilerindeki 2000 sonrasındaki kişilerin ülkelerini, yaşlarını görselleştirin.(Her bir Kategori için ayrı görselleştirme yapılması istenmektedir) Veriyi yorumlayınız.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/95onu/Downloads/asset-v1_Kodlasam+MS01+2023_06+type@asset+block@nobel.csv')
```

```python
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

```python
df_2000_sonrasi = df[df['year'].dt.year > 2000]
```

```python
def plot_age_distribution(category, country_column):
    category_data = df_2000_sonrasi[df_2000_sonrasi['category'] == category]
    
    if category_data.empty:
        print(f"{category} kategorisinde veri bulunamadı.")
        return
    
    category_data = category_data.copy()

    category_data.loc[:, 'age'] = category_data['year'].dt.year - pd.to_datetime(category_data['birth_date']).dt.year
    
    plt.figure(figsize=(12, 6))
    sns.barplot(x=country_column, y='age', data=category_data, palette='viridis')
    plt.xticks(rotation=90)
    plt.xlabel('Ülke')
    plt.ylabel('Yaş')
    plt.title(f'{category} Kategorisinde 2000 Sonrası Nobel Ödülü Alanların Yaş Dağılımı')
    plt.show()
```

```python
categories = ['Chemistry', 'Literature', 'Peace', 'Physics', 'Medicine']
for category in categories:
    if category in ['Peace', 'Literature']:
        plot_age_distribution(category, 'birth_country')
    else:
        plot_age_distribution(category, 'organization_country')
```
